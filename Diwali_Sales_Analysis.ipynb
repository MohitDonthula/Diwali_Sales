```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
df.dropna(inplace = True)
```

```python
df.dtypes
```

```python
df['Amount'] = df['Amount'].astype('int64')
```

```python
df.dtypes
```

```python
df.rename(columns = {'Gender' : 'Sex'})
```

```python
ax = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = ax)
```

```python
state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head()

plt.figure(figsize=(10, 5))
sns.barplot(x = 'State', y = 'Orders', data = state)
```

```python
amount_gen = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(7)

plt.figure(figsize = (15,5))
sns.barplot(x = 'State', y = 'Amount', data = amount_gen)
```

```python
amount_byMarital_status = df.groupby(['Marital_Status'], as_index = False)['Amount'].sum()

sns.barplot(x = 'Marital_Status', y = 'Amount', data = amount_byMarital_status )
```
